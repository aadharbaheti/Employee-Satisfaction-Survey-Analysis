```python
from scipy.stats import f_oneway

low_salary = data[data['salary'] == 'low']['average_montly_hours']
medium_salary = data[data['salary'] == 'medium']['average_montly_hours']
high_salary = data[data['salary'] == 'high']['average_montly_hours']

f_statistic, p_value = f_oneway(low_salary, medium_salary, high_salary)

if p_value < 0.05:
    print(f"The one-way ANOVA test for 'average_monthly_hours' is statistically significant (p-value={p_value}).")
else:
    print(f"The one-way ANOVA test for 'average_monthly_hours' is not statistically significant (p-value={p_value}).")
```

```python
from scipy.stats import f_oneway

# Extract the 'number_project' data for each salary category
low_salary_projects = data[data['salary'] == 'low']['number_project']
medium_salary_projects = data[data['salary'] == 'medium']['number_project']
high_salary_projects = data[data['salary'] == 'high']['number_project']

# Perform one-way ANOVA test
f_statistic, p_value = f_oneway(low_salary_projects, medium_salary_projects, high_salary_projects)

if p_value < 0.05:
    print(f"The one-way ANOVA test for 'number_project' is statistically significant (p-value={p_value}).")
else:
    print(f"The one-way ANOVA test for 'number_project' is not statistically significant (p-value={p_value}).")
```

```python
from scipy.stats import f_oneway

# Extract 'last_evaluation' data for each salary category
low_salary_evaluation = data[data['salary'] == 'low']['last_evaluation']
medium_salary_evaluation = data[data['salary'] == 'medium']['last_evaluation']
high_salary_evaluation = data[data['salary'] == 'high']['last_evaluation']

# Perform one-way ANOVA test
f_statistic, p_value = f_oneway(low_salary_evaluation, medium_salary_evaluation, high_salary_evaluation)

if p_value < 0.05:
    print(f"The one-way ANOVA test for 'last_evaluation' is statistically significant (p-value={p_value}).")
else:
    print(f"The one-way ANOVA test for 'last_evaluation' is not statistically significant (p-value={p_value}).")
```

```python
# Create a contingency table for 'salary' and 'dept'
contingency_table = pd.crosstab(data['salary'], data['dept'])


# Create a clustered bar chart
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Clustered Bar Chart of Salary vs. Department')
plt.xlabel('Salary Category')
plt.ylabel('Count')
plt.legend(title='Department', loc='upper right')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.show()
```

```python
from scipy.stats import kruskal

# Extract satisfaction level data for each salary category
low_salary_satisfaction = data[data['salary'] == 'low']['satisfaction_level']
medium_salary_satisfaction = data[data['salary'] == 'medium']['satisfaction_level']
high_salary_satisfaction = data[data['salary'] == 'high']['satisfaction_level']

# Perform the Kruskal-Wallis test
statistic, p_value = kruskal(low_salary_satisfaction, medium_salary_satisfaction, high_salary_satisfaction)

if p_value < 0.05:
    print(f"The Kruskal-Wallis test is statistically significant (p-value={p_value}).")
    print("There is a significant relationship between 'salary' and 'satisfaction_level'.")
else:
    print(f"The Kruskal-Wallis test is not statistically significant (p-value={p_value}).")
    print("There is no significant relationship between 'salary' and 'satisfaction_level'.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get feature importances from the model
feature_importances = model.feature_importances_

# Get the names of the features (column names of X)
feature_names = X.columns

# Sort the features by importance in descending order
sorted_indices = np.argsort(feature_importances)[::-1]
sorted_feature_importances = feature_importances[sorted_indices]
sorted_feature_names = feature_names[sorted_indices]

# Create a bar plot of feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(len(sorted_feature_importances)), sorted_feature_importances)
plt.xticks(range(len(sorted_feature_importances)), sorted_feature_names, rotation=90)
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Feature Importance")
plt.grid(True)
plt.show()
```
